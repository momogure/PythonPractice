```python
"""
Bagles

"Pico" --> a correct digit in the wrong place
"Fermi" --> a correct digit in the correct place
"Bagels" --> no correct digits

"""
import random

pico = "Pico"
fermi = "Fermi"
bagles = "Bagels"

N_TRIAL = 10
digits = 5

# Check if theere are duplicate element in the list
def check_duplicates(my_list):
    n_dupes = 0
    dummy_list = [1 for x in my_list if my_list.count(x) > 1]
    for n in dummy_list:
        n_dupes += n
    return n_dupes

# Check if my_num is interger
def check_number(my_num):
    my_num = [isinstance(int(n), int) for n in my_num]
    result = 0
    for num in my_num:
        result += num
        return result
       
# Obtain your guess: 'your_guess'
def get_guess():
    print('Enter {} digit number'.format(digits))
    your_guess = input()
    while not(check_number(your_guess)):
        print('Enter {} digit number'.format(digits))
        your_guess = input()
    guess = [int(n) for n in your_guess]
    return guess

# Create a random number with no duplicates: 'num' --> Ex. these are no good 012, 499
num = [random.randint(0, 9) for n in range(digits)]
has_duplicates = check_duplicates(num)

# If the list num contains duplicate numbers or the first element is 0, repeats to generate num
while has_duplicates | (num[0] == 0):
    num = [random.randint(0, 9) for n in range(digits)]

# Obtain the number
your_guess = get_guess()

i = 0 # Count the number of try
answer = [bagles for x in range(digits)]  # Initialize 'answer' that stores 'pico', 'fermi', or 'bagles'
bingo = [fermi for x in range(digits)]  # If 'snswer' equals to bingo, you win

while (i < N_TRIAL) & (answer != bingo):
    # Loop each digit of the randomly generate number
    for j in range(len(num)):
        # Loop each digit of the user input number
        for k in range(len(your_guess)):
            if (num[j] == your_guess[k]): # Both numbers match
                if j == k: # The positions are the same
                    answer[k] = fermi
                else: # The positions are different
                    answer[k] = pico
                k = 0
                j += 1
                break
            else:
                k += 1
        j += 1        
    i += 1
    print("Answer: {}".format(answer))
    if answer == bingo:
        break
    your_guess = get_guess()

print("Bingo! you guessed {}!".format(num))
```
